## Combining Two Different of Plots

```python
# Data for the bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 15, 7, 12, 5]

# Data for the line plot
x = np.arange(len(categories))
line_values = [5, 12, 9, 7, 8]
```

```python
plt.bar(categories, values)
plt.plot(x, line_values, color='r')
plt.show()
```

```python
# Create the figure and axis objects
fig, ax1 = plt.subplots()

# Bar chart on the left y-axis
ax1.bar(categories, values, color='lightblue', label='Bar Chart')

# Create a second y-axis to plot the line chart
ax2 = ax1.twinx()
ax2.plot(x, line_values, color='red', marker='o', label='Line Plot', linestyle='-', linewidth=2)
# Add labels and title
ax1.set_xlabel('Categories')
ax1.set_ylabel('Bar Chart Values', color='blue')
ax2.set_ylabel('Line Chart Values', color='red')
plt.title('Combined Bar Chart and Line Plot')

# Adding legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Display the plot
plt.show()
```

# Sub-plots

```python
# Generate random data for four scatter plots
data = [np.random.rand(2, 10) for _ in range(4)]
data
```

```python
# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

print(f"axs: {axs}")

# Titles for each subplot
titles = ["Scatter Plot 1", "Scatter Plot 2", "Scatter Plot 3", "Scatter Plot 4"]

# Plot each dataset in a separate subplot
for i, ax in enumerate(axs.flat):
    ax.scatter(data[i][0], data[i][1])
    ax.set_title(titles[i])
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

### Class Assignment

```python
np.random.seed(42)
scatter_data = np.random.rand(2, 15)
scatter_sizes = np.random.randint(80, 200, (15,))
scatter_colors = np.random.rand(15) 
```

```python
bar_y = np.random.randint(-10, 10, (5,))
bar_x = ['A', 'B', 'C', 'D', 'E']
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

axs[0].scatter(scatter_data[0], scatter_data[1], c=scatter_colors, sizes=scatter_sizes, cmap='viridis', label="Scatter")
axs[0].set_title('Scatter of random samples')
axs[0].set_xlabel('Random X Label')
axs[0].set_ylabel('Random Y Label')

# profit_color = [{p<0: 'red', 0<=p<=2: 'orange', p>2: 'green'}[True] for p in bar_y]
profit_color = list(map(lambda x: 'red' if x<0 else 'orange' if 0<=x<=2 else 'green', bar_y))
bars = axs[1].bar(bar_x, bar_y, color=profit_color, label='Bar')
axs[1].set_title('Bar Chart of random samples')
axs[1].set_xlabel('Random X Categories')
axs[1].set_ylabel('Random Y Values')

for i, bar in enumerate(bars):
    yval = bar.get_height()
    y_pos = yval + .05 if profit_color[i] != 'red' else yval - .5
    axs[1].text(bar.get_x() + .3,y_pos , yval)
    
fig.supxlabel('common_x')
fig.supylabel('common_y')
# fig.tight_layout()
fig.legend()
plt.show()
```
