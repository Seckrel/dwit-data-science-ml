```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
```

![3D Data Cube](./images/sv_graph.png "Example of 3D Matrix")  

![3D Data Cube](./images/svm_kernal.png "Example of 3D Matrix")  

```python
X, y = datasets.make_circles(n_samples=100, factor=0.5, noise=0.1)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, s=50, edgecolors='k', cmap=plt.cm.Set3)
plt.show()
```

```python
svm_model = SVC(kernel='rbf', gamma='scale')
svm_model = svm_model.fit(X_scaled, y)
```

```python
x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                     np.linspace(y_min, y_max, 100))
```

```python
Z = svm_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Set3)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, s=50, edgecolors='k', cmap=plt.cm.Set3)
plt.title('SVM with RBF Kernel - Decision Boundary')
plt.show()
```
