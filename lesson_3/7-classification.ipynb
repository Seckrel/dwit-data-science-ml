```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Loading Data

```python
data = load_breast_cancer()
feature_names = data.feature_names
df = pd.DataFrame(data.data, columns=feature_names)
```

```python
df.head()
```

```python
target_names = data.target_names
df['target'] = data.target
df['target_name'] = pd.Categorical.from_codes(data.target, target_names)
df.head()
```

```python
df.drop(columns=['target']).describe()
```

# EDA

## Corr Heatmap

```python
import seaborn as sns
```

```python
corr_matrix = df.drop(columns=['target', 'target_name']).corr()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False, linewidth=.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

## Pair Plot

```python
sns.pairplot(df, vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area'], hue='target', diag_kind='kde')
plt.suptitle("Pair Plot of Selected Features", y=1.02)
plt.show()
```

# Classifier GNB

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[feature_names], df['target'], test_size=.3, random_state=42)
```

```python
sns.countplot(x=y_train, legend=False)
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf = GaussianNB()
clf.fit(X_train, y_train)
```

## Metrics

```python
from sklearn.metrics import precision_score, recall_score, roc_curve, auc, ConfusionMatrixDisplay, f1_score
```

```python
y_pred = clf.predict(X_test)
y_proba = clf.predict_proba(X_test)[:, 1]
```

### Confuction Matrix


```python
ConfusionMatrixDisplay.from_estimator(clf, X_test, y_test, display_labels=data.target_names, cmap='Blues')
plt.show()
```

### Precision
$$
P = \frac{TP}{TP + FP}
$$
>It allows us to look at accurarcy of positive prediction

```python
precision_score(y_pred, y_test)
```

### Recall
$$
R = \frac{TP}{TP + FN}
$$
>The Ratio of positive instance that are detected by the classifier
>Also called sensitivity or True Positive Rate

```python
recall_score(y_pred, y_test)
```

### F1-Score
$$
F1 = \frac{2 * precision * recall}{precision + recall}
$$

>Harmonic Mean of Recall and Precision
>It penalizes Outliers

```python
f1_score(y_pred, y_test)
```

### Precision vs. Recall
**When We Need High Precision**  
1. Focus on Accuracy of Positive Prediction
    - You want to minimize false positives (incorrreclty predicting positives).
    - Every positive prediction must be highly reliable because acting on a false positive can have significant costs or negative consequences.
- Trade off: Higher precision means some actual positives could be missed

**When We need High Recall**  
1. Focus on Identifying All Positives
    - You want to minimize false negatives (missing actual positive)
    - Missing a positive case has severe consequences, so you're willing to tolerate some false positives as long as most true positives are captured.

- Trade off: Higer recall means more false positive are accepted

$$
TPR = \frac{TP}{TP + FN}
$$
>TPR: from all people who has cancer, what percentage did we classify correctly  
>If all positives were correctly classified, TPR = 100%  
>If all positives were miss-classified, TPR = 0%

$$
FPR = \frac{FP}{FP + TN}
$$
>FPR: from all people who did not have cancer, what percentage did we miss-classify  
>If all negatives were missed, FPR = 100%  
>If all negatives were miss-classified, FPR = 0%

### ROC curve
- ROC Curve: Plots TPR (recall) against FPR.
- AUC: Measures the classifier's ability to distinguish between classes. An AUC of 1 indicates a perfect model.
- Insights
    - How thresholds affect model performance
    - Trade-offs between sensitivity and specificity
    - Comparision between classifiers (a higher AUC indicates a better model)

```python
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
roc_auc = auc(fpr, tpr)
```

```python
thresholds
```

```python
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})', color='blue', zorder=0)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Random classifier line

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')

plt.show()
```
